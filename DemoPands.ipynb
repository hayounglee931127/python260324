```python
#DemoPandas.ipynb
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame, Series

#시리즈연습
myCampus = Series([100, 200, 300, 400, 500], 
    index=['2026-03-01', '2026-03-02', '2026-03-03', '2026-03-04', '2026-03-05'])
print(myCampus)

#shift+enter : 실행
```

```python
#딕셔너리를 입력
data = {"foreigner":[1,2,3,4,5,6],
        "sratio":[10,20,30,40,50,60],
        "org":[100,200,300,400,500,600],
        "sprice":[1,2,3,4,5,6],
        "private":[10,20,30,40,50,60]
        }
data

frame = DataFrame(data=data)
frame

```

```python
# 컬럼 순서
frame2 = DataFrame(data=data, 
        columns=["foreigner", "org", "sprice", "private", "sratio"],
        index=["2026-03-01", "2026-03-02", "2026-03-03", "2026-03-04", "2026-03-05", "2026-03-06"]
        )
frame2
```

```python
#컬럼을 슬라이싱
frame2["private"]
```

```python
#특정행 슬라이싱
frame2.loc["2026-03-02"]
```

```python
#행열 전환 pivot
frame2.T
frame2.transpose()
```
